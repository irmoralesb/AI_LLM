# Audio Classification

## Steps

1. Transcribe audio
2. Detect language
3. Translate to English
   * Moderation
4. Generate response
   * Moderation
5. Replay in original language

## Case Study Plan

1. Transcribe the audio into text
2. Detect the language
3. Translate into English
4. Refining the text

```python
from openai import OpenAI 
from dotenv import load_dotenv

load_dotenv()
client = OpenAI()


def transcribe_audio(audio_file):
    return client.audio.transcriptions.create(
        model="whisper-1",
        file=audio_file
    )

def detect_language(text):
    response = client.chat.completions.create(
        model = "gpt-4o-mini",
        max_completion_tokens=5,
        messages=[
            {
                "role":"user",
                "content": f"""Identify the language of the following text and respond only with the country code
                 (e.g., 'en', 'uk', 'fr'): 
                 {text}"""
            }
        ]
    )
    return response.choices[0].message.content

def translate_to_english(country_code, text):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        max_completion_tokens=300,
        messages=[
            {
                "role":"user",
                "content":f"Translate this customer transcript from country code {country_code} to Englishm don't add any additional text: {text}"
            }
        ]
    )
    return response.choices[0].message.content

def refining_english_text(text):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        max_completion_tokens=300,
        messages = [
            {
                "role":"user",
                "content": f"""You are an AI assistant that corrects transcripts by fixing misinterpretations, names, and terminology.
                Please refine the following transcript:\n\n:{text}"""
            }
        ]
    )
    return response.choices[0].message.content
```

### Step 1: Transcribe Audio

```python
transcript = ""
with open("../resources/customer_call_it_v2.wav", "rb") as audio_file:
    audio_transcribed = transcribe_audio(audio_file)
    transcript = audio_transcribed.text
    print(transcript)
```

## Step 2: Detect Language

```python
detected_language = detect_language(transcript)
print(detected_language)
```

### Step 3: Translate to English

```python
transcript_in_english = translate_to_english(detect_language, transcript)
print(transcript_in_english)
```

### Step 4: Refining the Text


```python
transcript_in_english_refined = refining_english_text(transcript_in_english)
print(transcript_in_english_refined)
```
