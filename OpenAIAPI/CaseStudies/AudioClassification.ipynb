# Audio Classification

## Steps

1. Transcribe audio
2. Detect language
3. Translate to English
   * Moderation
4. Generate response
   * Moderation
5. Replay in original language

## Case Study Plan

1. Transcribe the audio into text
2. Detect the language
3. Translate into English
4. Refining the text

```python
from audio_classification_utils import *
```

### Step 1: Transcribe Audio

```python
transcript = ""
with open("../resources/customer_call_it_v2.wav", "rb") as audio_file:
    audio_transcribed = transcribe_audio(audio_file)
    transcript = audio_transcribed.text
    print(transcript)
```

## Step 2: Detect Language

```python
detected_language = detect_language(transcript)
print(detected_language)
```

### Step 3: Translate to English

```python
transcript_in_english = translate_to_english(detected_language, transcript)
print(transcript_in_english)
```

### Step 4: Refining the Text


```python
transcript_in_english_refined = refining_english_text(transcript_in_english)
print(transcript_in_english_refined)
```

### Step 5: Moderation

```python
is_moderated = validate_transcript_is_moderated(transcript_in_english_refined)
print("It's Moderated" if is_moderated else "It's not Moderated")
```

### Step 6: Generating Response

```python
if is_moderated:
    transcript_response = generate_transcript_response(transcript_in_english_refined)
    print (transcript_response)
```

### Step 7: Moderation (for the response)

```python
if is_moderated:
    is_moderated = validate_transcript_is_moderated(transcript_response)
    print("It's Moderated" if is_moderated else "It's not Moderated")
```

### Step 8: Translating the Transcription Response


```python
if is_moderated:
    translated_response = translate_to_original_language(detected_language, transcript_response)
    print(translated_response)
```

### Step 9: Generating Speech Response

```python
output_file_name = text_to_speech(translated_response)
```
