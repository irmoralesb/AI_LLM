## How to give extra information?

* Sample previous conversations
* System's prompts  

### Sample Conversations

* Guide model to answer specific questions

##### Disadvantage: might need a lot of samples  

````python
response = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {'role':'system', 'content':'You are a customer service chatbot that responds to user queries in a gentle way.'},
        {'role':'user', 'content':'What services do you offer?'},
        {'role':'assistant', 'content':'We provide services for web application development, mobile app development, and custom software solutions'},
        {'role':'user', 'content':'How many services do you have?'}
    ]
)

print(response.choices[0].message.content)
````

### System prompt

* Includes required context for chatbot answers

````python
services = "ABC Tech Solution, a leading IT company, offers a range of services: application development, mobile app development, and custom software solutions."
system_prompt = f"""
You are a customer service chatbot that responds to user queries in a gentle way. Some information about our services are delimited by triple backticks.

```{services}```
"""

user_prompt = "How many services do you offer?"
print(get_response(system_prompt, user_prompt))
````
