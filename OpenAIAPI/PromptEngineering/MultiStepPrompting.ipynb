# Multi-step Prompting

* Break down an end goal into series of steps
* Model goes through each step to give final output
* Multi-step prompts are used for:
   * Sequential tasks
   * Cognotive tasks


## Multi-step vs Few-shot prompt

Steps:
* Explicitly tell model what to do


Shots: 
* Questions and Answers model learns from

### Analizing solution correctness

* Checking solution correctness requires multiple steps

Example:
* Python code for calculation functions

````python
calculator = """
def add(a, b):
    return a + b

def substract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b
"""
````

Using Single-Step Prompt

````python
prompt = f"""Determine if the code delimited by triple backticks is correct or not.
Answer by yes or no.
```{calculator}```
"""

print_response(prompt)
````

Using Multi-Step Prompt

````python
prompt = f"""Determine the correctness of the code delimited by triple backticks as follows:
Step 1: Check the code correctness in each function.
Step 2: Verify if the divide function handles the case when dividing by 0.

Code: ```{calculator}```
"""

print_response(prompt)
````
