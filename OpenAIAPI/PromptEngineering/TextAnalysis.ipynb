### Entity Extraction with Few-Shot Prompts

* For complex structures

````python
ticket_1 = """
Hello, I'm Emma Adams. I'd like to ask about my reservation with the code CAR123.
You can reach me at +123456 if needed.
"""

ticket_2 = """
This is Sarah Williams.
I would like to request some information regarding my upcoming flight with reservation code FLIGHT987. Thank you.
"""

````

> We may want to get the entity data as follows

````python
entities_1 = """ 
* Customer Details:
    - Name: Emma Adams
    - Phone: +123456
* Reservation Details
    - Reservation Code: CAR123
"""
entities_2 = """
* Customer Details:
    - Name: Sara Williams
* Reservation Details
    - Reservation Code: FLIGHT987
"""
````

> Now we want to get the entities from a third ticket

````python
ticket_3 = """
Hello, I'm David Brown (CUST123). 
I need assistance with my reservation under the code HOTEL456.
There are some questions and issues related to my upcoming stay that require your attention.
"""

prompt =f"""
Ticket: {ticket_1} -> Entities: {entities_1}
Ticket: {ticket_2} -> Entities: {entities_2}
Ticket: {ticket_3} -> Entities:
"""

print(get_response(prompt))
````
