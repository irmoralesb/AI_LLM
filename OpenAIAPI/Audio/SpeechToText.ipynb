# Speech to Text

Speech to text capabilities

* Transcribe audio
* Translate non-English audio
* Support mp3, mp4, mpeg, mpga, m4a, wav, and webm.
   * 25 MB limit

## Use cases

* Meeting transcripts
* Video captions
* Processing customer calls

## Transcription Performance

* Performance can vary widly, depending on:
   * Audio quality
   * Audio language
   * Model's knowledge of the subject matter

```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI()
```

## Transcripts

```python
def transcript_audio(audio_file):
    response = client.audio.transcriptions.create(
        model="whisper-1",
        file=audio_file
    )
    return response

with open("../audio_files/audio-logan-advocate-openai.mp3","rb") as audio_file:
    transcription = transcript_audio(audio_file)
    print(transcription.text)
```

### Transcription in Portuguese

```python
with open("../audio_files/audio-portuguese.m4a","rb") as audio_file:
    transcription = transcript_audio(audio_file)
    print(transcription.text)
```

## Translations

```python
def translate_audio(audio_file):
    response = client.audio.translations.create(
        model="whisper-1",
        file=audio_file
    )
    return response

with open("../audio_files/audio-portuguese.m4a","rb") as audio_file:
    translation = translate_audio(audio_file)
    print(translation.text)
```
